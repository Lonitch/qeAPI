## DOS/PDOS Visualization

```python
from qeplot import *
from qe2cif import *
import matplotlib.pyplot as plt
%matplotlib inline
dospath = 'C://Users//liu_s//Documents//BW_results//SA//Naf1//GS//DOS-PDOS'
# the output file of NSCF calculation to get fermi energy level
nscfpath = 'C://Users//liu_s//Documents//BW_results//SA//Naf1//GS//DOS-PDOS'
```

```python
plt.rcParams['font.size']=18
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
datafile = 'C://Users//liu_s//Documents//BW_results//SA//Naf1//GS//SOC//VLAB.pdos_tot'
dosplot(ax,datafile=datafile,xrange=5,fermi=2.8708,lw=2)
ax.set_title('Naf1(noncolin+SOC)')
```
