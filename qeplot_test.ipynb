## DOS/PDOS Visualization

```python
from qeplot import *
from qe2cif import *
import matplotlib.pyplot as plt
%matplotlib inline
dospath = 'C://Users//liu_s//Documents//BW_results//SA//Naf1//GS//DOS-PDOS'
# the output file of NSCF calculation to get fermi energy level
nscfpath = 'C://Users//liu_s//Documents//BW_results//SA//Naf1//GS//DOS-PDOS'
```

```python
plt.rcParams['font.size']=18
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
dosplot(dospath,ax,xrange=10)
```

```python
pdospath = 'C://Users//liu_s//Documents//BW_results//SA//Naf1//GS//DOS-PDOS'
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
res,fermiE=pdosplot(pdospath,ax,nscfpath=nscfpath,orbital='by element',xrange=15)
ax.set_ylim([-10,10])
plt.legend(loc=4, prop={'size': 12})
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
res2,fermiE=pdosplot(pdospath,ax,nscfpath=nscfpath,orbital='by orbital',xrange=15)
ax.set_ylim([-10,10])
plt.legend(loc=4, prop={'size': 12})
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
res3,fermiE=pdosplot(pdospath,ax,nscfpath=nscfpath,orbital=[('Fe','3d'),('Ni','3d'),('Na','2p'),('O','2p')],xrange=15)
ax.set_ylim([-10,10])
plt.legend(loc=4, prop={'size': 12})
```
